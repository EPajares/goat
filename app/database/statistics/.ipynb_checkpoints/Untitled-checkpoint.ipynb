```python
#Basic statistical variables
n = np.size(data['accessibility_index'])
max_ai = max(data['accessibility_index'])
min_ai = min(data['accessibility_index'])
mean = np.nanmean(data['accessibility_index'])
median = np.nanmedian(data['accessibility_index'])
standard_deviation = np.nanstd(data['accessibility_index'])
variance = np.nanvar(data['accessibility_index'])
print(n, max_ai, min_ai, mean, median, standard_deviation, variance)
```

```python
#Range -> 0 is perfect equality, n is perfect inequality (one person recieves all)
E = (max_ai-min_ai)/mean
#Relative Mean Deviation
M = sum(np.absolute(mean-data['accessibility_index']))*n*mean
#Mean weighted by population
Mean_weighted = (np.sum(data['accessibility_index']*data['population']))/np.sum(data['population'])
#Variance and Coefficient Variation
#V = sum((mean-data['accessibility_index'])^2)/n
E, M, Mean_weighted
```
