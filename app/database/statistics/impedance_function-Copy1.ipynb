```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.optimize import curve_fit
from scipy import optimize
from lmfit.models import StepModel, LinearModel
from sklearn.metrics import r2_score
```

```python
def gaus(x, b):
    return (np.exp(-((x)**2)/b))
def exp(x, b):
    return(np.exp(x*b))
def exp2(x, a, b):
    return(a*np.exp(b*x))
def pow(x, a):
    return(x**(-a))
```

```python
#Daily Goods
x = np.array([375, 750, 1500, 3750])
x2 = np.array([187.5,562.5,1125,2625])
y = np.array([0.61, 0.29, 0.08, 0.02])

#Gaussian Function
beta1 = optimize.curve_fit(gaus, xdata = x2, ydata = y, p0 = 40000)[0]
print('Gaussian Function')
print(f'beta = {beta1[0]}')
print('R2 = ', r2_score(y, gaus(x2,*beta1)))

#Exponential Function 1
alpha2 = optimize.curve_fit(exp, xdata = x2, ydata = y, p0 = 0.01)[0]
print('Exponential Function 1')
print(f'alpha = {alpha2[0]}')
print('R2 = ', r2_score(y, exp(x2,*alpha2)))

#Exponential Function 2
alpha3, beta3 = optimize.curve_fit(exp2, xdata = x2, ydata = y, p0 = (5.0,0.01))
alpha3a = alpha3[0]
print('Exponential Function 2')
print(f'alpha = {alpha3[0]}', f'beta = {alpha3[1]}')
print('R2 = ', r2_score(y, exp2(x2,alpha3[0],alpha3[1])))

#Power Function
alpha4 = optimize.curve_fit(pow, xdata = x2, ydata = y, p0 = (0.4))
print('Power Function')
print(f'alpha = {alpha4[0][0]}')
print('R2 = ', r2_score(y, pow(x2,alpha4[0])))

#plot Graphs
xmin, xmax = 0, 6000
a = plt.figure(num=1,dpi=500)
points = 1000
axes = a.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlim([0,6000])
axes.set_ylim([0,1])
x2 = np.array([187.5,562.5,1125,2625])
y = np.array([0.61, 0.29, 0.08, 0.02])
#plt.step(x1, y, where='pre')
plt.plot(x2, y, 'o--', color='grey', alpha=0.3)
plt.title(label='Daily Goods')
axes.set_xlabel('Time [s]')
axes.set_ylabel('Percentage of Accessibility')
axes.grid(True, color ='#EEEEEE')
x = np.linspace(xmin, xmax, points)
plt.plot(x, gaus(x, beta1), label = 'Modified Gaussian')
plt.plot(x, exp(x, alpha2), label = 'Exponential 1')
plt.plot(x, exp2(x, alpha3[0], alpha3[1]), label = 'Exponential 2')
plt.plot(x, pow(x, alpha4[0][0]), label = 'Power')
plt.legend(loc='upper right')
plt.show()
```
